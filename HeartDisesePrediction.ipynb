```python
X = heart_data.drop(columns='HeartDisease', axis=1)
Y = heart_data['HeartDisease']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the Data into Training data & Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
# Step 1: Import the necessary libraries
from sklearn.preprocessing import StandardScaler

# Step 2: Feature Scaling (Important for Logistic Regression)
scaler = StandardScaler()  # Create a StandardScaler object
X_train_scaled = scaler.fit_transform(X_train)  # Fit and transform on training data
X_test_scaled = scaler.transform(X_test)  # Only transform test data (not fitting again)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Step 1: Handle missing values using SimpleImputer (impute with the mean)
imputer = SimpleImputer(strategy='mean')  # 'median' or 'most_frequent' are also valid strategies
X_train_imputed = imputer.fit_transform(X_train)  # Fit and transform on training data
X_test_imputed = imputer.transform(X_test)  # Transform test data with the same imputer

# Step 2: Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_imputed)  # Fit and transform on training data
X_test_scaled = scaler.transform(X_test_imputed)  # Transform test data (not fitting again)

# Step 3: Train the Logistic Regression model
logreg = LogisticRegression(random_state=2)
logreg.fit(X_train_scaled, Y_train)  # Train the model on scaled, imputed training data

# Step 4: Make predictions and evaluate the model
Y_pred = logreg.predict(X_test_scaled)
print("Accuracy:", accuracy_score(Y_test, Y_pred))  # Evaluate the model using accuracy
```

Model Evaluation

Accuracy Score

```python
# Accuracy on training data
X_train_prediction = logreg.predict(X_train_scaled)  # Predict on the training data
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)  # Accuracy on the training data
print("Training Data Accuracy:", training_data_accuracy)
```

```python
# Accuracy on test data
X_test_prediction = logreg.predict(X_test_scaled)  # Predict on the test data
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)  # Accuracy on the test data
print("Test Data Accuracy:", test_data_accuracy)
```

Building a Predictive System

```python
input_data = (37,0,0,180,260,0,1,187,0,1,1)

# (Age,Sex,ChestPainType,RestingBP,Cholesterol,FastingBS,RestingECG,MaxHR,ExerciseAngina,Oldpeak,ST_Slope,HeartDisease)

# change the input data to a numpy array
input_data_as_numpy_array= np.asarray(input_data)

# reshape the numpy array as we are predicting for only on instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = logreg.predict(input_data_reshaped)
print(prediction)

if (prediction[0]== 0):
  print('The Person does not have a Heart Disease')
else:
  print('The Person has Heart Disease')
```
